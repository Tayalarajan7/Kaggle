```python
# Importing required libraries and modules
# Data manipulation and analysis libraries
import pandas as pd
import numpy as np
# Visualization libraries
import seaborn as sns
import matplotlib.pylab as plt
%matplotlib inline
# Sklearn modules
from sklearn.datasets import make_classification  
from sklearn.model_selection import train_test_split, cross_val_score, GridSearchCV
from sklearn.linear_model import LogisticRegression
from sklearn.neighbors import KNeighborsClassifier
from sklearn.metrics import accuracy_score, classification_report
from sklearn.preprocessing import StandardScaler, LabelEncoder, StandardScaler
from sklearn.ensemble import RandomForestClassifier
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.cluster import KMeans
from sklearn.decomposition import PCA
# TensorFlow and Keras modules
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense, Dropout, Input
from tensorflow.keras.utils import to_categorical
from keras.callbacks import EarlyStopping
```

```python
#Reading the data from .csv file
shark_tank_df = pd.read_csv('Shark Tank Australia dataset.csv')
```

```python
#Displaying the dataframe
shark_tank_df.head()
```

```python
# Display basic information about the DataFrame
print("Data Information:")
print(shark_tank_df.info())
```

```python
# List of columns with constant values
constant_columns = [
    'Royalty Deal', 'Deal Has Conditions', 'Steve Present',
    'Janine Present', 'Andrew Present', 'Naomi Present', 'Glen Present', 'Advisory Shares Equity'
]

# Remove constant columns from the DataFrame
shark_tank_df.drop(columns=constant_columns, inplace=True)

# Display the remaining columns in the DataFrame
print("Remaining columns in the DataFrame after removing constant columns:")
print(shark_tank_df.columns)
```

```python
# Display the remaining columns in the DataFrame
print("Remaining columns in the DataFrame:")
print(shark_tank_df.columns)
```

```python
# Checking for missing values in each column
print("Count of missing values in each column before replacement:")
print(shark_tank_df.isnull().sum())
```

```python
# Replace missing values in numerical columns with the mean
numerical_columns = shark_tank_df.select_dtypes(include=[np.number]).columns
for column in numerical_columns:
    shark_tank_df[column].fillna(shark_tank_df[column].mean(), inplace=True)

# Replace missing values in categorical columns with the mode
categorical_columns = shark_tank_df.select_dtypes(include=[object]).columns
for column in categorical_columns:
    shark_tank_df[column].fillna(shark_tank_df[column].mode()[0], inplace=True)

# Checking for missing values in each column after replacement
print("Count of missing values in each column after replacement:")
print(shark_tank_df.isnull().sum())
```

```python
# Filter the DataFrame to include only the numerical columns
shark_tank_numerical_df = shark_tank_df[numerical_columns]
# Calculate the interquartile range (IQR) for each numerical column
Q1 = shark_tank_numerical_df.quantile(0.25)
Q3 = shark_tank_numerical_df.quantile(0.75)
IQR = Q3 - Q1
# Determine the threshold for outliers
outlier_threshold = 1.5
# Identify rows containing outliers in any numerical column
outliers = ((shark_tank_numerical_df < (Q1 - outlier_threshold * IQR)) | (shark_tank_numerical_df > (Q3 + outlier_threshold * IQR)))
# Count the number of outliers in each numerical column
outliers_count=outliers.sum()

print("Outliers count in each numerical columns:")
print(outliers_count)
```

```python
# Remove rows containing outliers
cleaned_shark_tank_df = shark_tank_df[~outliers.any(axis=1)]

# Display the shape of the original and cleaned DataFrames
print("Original DataFrame shape:", shark_tank_df.shape)
print("Cleaned DataFrame shape:", cleaned_shark_tank_df.shape)
```

```python
# Group by 'Industry' and count the number of pitches in each industry
industry_pitches = shark_tank_df['Industry'].value_counts()

# Print the industries with the most pitches
print("Industries with the most pitches:")
print(industry_pitches)
```

```python
# Define a color palette
color_palette = plt.get_cmap('tab20')
colors = {industry: color_palette(i) for i, industry in enumerate(industry_pitches.index)}

# Plot the data as a pie chart
plt.figure(figsize=(10, 8))
industry_pitches.plot(kind='pie', autopct='%1.1f%%', startangle=180, legend=False, colors=[colors[ind] for ind in industry_pitches.index])
plt.title('Number of Pitches per Industry')
plt.ylabel('')  # Hide the y-label for better visualization
plt.show()

# Group by 'Season Number' and 'Industry' to see trends over different seasons
season_industry_trends = shark_tank_df.groupby(['Season Number', 'Industry']).size().unstack(fill_value=0)

# Plot the trends
plt.figure(figsize=(14, 8))
season_industry_trends.plot(kind='bar', stacked=True, figsize=(14, 8), color=[colors[ind] for ind in season_industry_trends.columns])
plt.title('Number of Pitches per Industry over Different Seasons')
plt.xlabel('Season Number')
plt.ylabel('Number of Pitches')
plt.legend(title='Industry', bbox_to_anchor=(1.05, 1), loc='upper left')
plt.tight_layout()
plt.show()
```

```python
# Calculate the overall success rate
total_pitches = shark_tank_df.shape[0]
successful_pitches = shark_tank_df[shark_tank_df['Received Offer'] == 1].shape[0]
overall_success_rate = (successful_pitches / total_pitches) * 100

print(f"Overall Success Rate: {overall_success_rate:.2f}%")
```

```python
# Calculate success rates by industry
industry_success = shark_tank_df.groupby('Industry')['Received Offer'].value_counts(normalize=True).unstack().fillna(0) * 100
industry_success = industry_success[1]

print("Success Rates by Industry:")
print(industry_success)
```

```python
plt.figure(figsize=(12, 6))
industry_success.sort_values().plot(kind='barh', color='cornflowerblue')
plt.title('Success Rates by Industry')
plt.xlabel('Success Rate (%)')
plt.ylabel('Industry')
plt.show()
```

```python
# Calculate success rates by season
season_success = shark_tank_df.groupby('Season Number')['Received Offer'].value_counts(normalize=True).unstack().fillna(0) * 100
season_success = season_success[1]

print("Success Rates by Season:")
print(season_success)
```

```python
# Plot success rates by season
plt.figure(figsize=(12, 6))
season_success.plot(kind='bar', color='mediumseagreen')
plt.title('Success Rates by Season')
plt.xlabel('Season Number')
plt.ylabel('Success Rate (%)')
plt.show()
```

```python
# Calculate average investment amounts by industry
industry_avg_investment = shark_tank_df.groupby('Industry')['Total Deal Amount'].mean().sort_values()

# Calculate average investment amounts by shark
# Assuming there are columns for each shark's investment amounts
shark_columns = [
    'Steve Investment Amount', 'Janine Investment Amount', 'Andrew Investment Amount',
    'Naomi Investment Amount', 'Glen Investment Amount', 'Guest Investment Amount'
]

# Calculate the total investment by each shark and then find the average
shark_avg_investment = shark_tank_df[shark_columns].mean().sort_values()

# Print the results
print("Average Investment Amounts by Industry:")
print(industry_avg_investment)
```

```python
print("\nAverage Investment Amounts by Shark:")
print(shark_avg_investment)
```

```python
# Plot average investment amounts by industry
plt.figure(figsize=(12, 6))
industry_avg_investment.plot(kind='barh', color='cornflowerblue')
plt.title('Average Investment Amounts by Industry')
plt.xlabel('Average Investment Amount ($)')
plt.ylabel('Industry')
plt.show()

# Plot average investment amounts by shark
plt.figure(figsize=(12, 6))
shark_avg_investment.plot(kind='barh', color='mediumseagreen')
plt.title('Average Investment Amounts by Shark')
plt.xlabel('Average Investment Amount ($)')
plt.ylabel('Shark')
plt.show()
```

```python
# Convert 'Received Offer' to binary (0 or 1)
shark_tank_df['Received Offer'] = shark_tank_df['Received Offer'].apply(lambda x: 1 if x == 'Yes' else 0)

# Ensure 'Business Description' column is of string type
shark_tank_df['Business Description'] = shark_tank_df['Business Description'].astype(str)

# Vectorize descriptions using TF-IDF
tfidf_vectorizer = TfidfVectorizer(stop_words='english')
tfidf_matrix = tfidf_vectorizer.fit_transform(shark_tank_df['Business Description'])

# Convert TF-IDF matrix to DataFrame
tfidf_df = pd.DataFrame(tfidf_matrix.toarray(), index=shark_tank_df.index)

# Select numeric columns for clustering
numeric_columns = ['Total Deal Amount', 'Total Deal Equity', 'Received Offer']
numeric_df = shark_tank_df[numeric_columns]

# Combine TF-IDF features with numeric features
combined_df = pd.concat([tfidf_df, numeric_df], axis=1)

# Convert all column names to strings
combined_df.columns = combined_df.columns.astype(str)

# Standardize the combined features
scaler = StandardScaler()
scaled_features = scaler.fit_transform(combined_df)

# Apply K-Means Clustering
kmeans = KMeans(n_clusters=5, random_state=42)
kmeans.fit(scaled_features)
shark_tank_df['Cluster'] = kmeans.labels_

# Visualize the clusters using PCA
pca = PCA(n_components=2)
pca_components = pca.fit_transform(scaled_features)

# Plot PCA components with clusters
plt.figure(figsize=(10, 8))
sns.scatterplot(x=pca_components[:, 0], y=pca_components[:, 1], hue=shark_tank_df['Cluster'], palette='Set1', s=100)
plt.title('Startup Clustering (PCA)')
plt.xlabel('PCA Component 1')
plt.ylabel('PCA Component 2')
plt.show()

# Visualize clusters in terms of original features
plt.figure(figsize=(10, 6))
sns.countplot(x='Industry', hue='Cluster', data=shark_tank_df, palette='Set1')
plt.title('Number of Startups per Industry by Cluster')
plt.xticks(rotation=90)
plt.show()
```

```python
# Select relevant features for shark clustering
shark_columns = [
    'Steve Investment Amount', 'Janine Investment Amount', 'Andrew Investment Amount',
    'Naomi Investment Amount', 'Glen Investment Amount', 'Guest Investment Amount',
    'Steve Investment Equity', 'Janine Investment Equity', 'Andrew Investment Equity',
    'Naomi Investment Equity', 'Glen Investment Equity', 'Guest Investment Equity'
]

shark_df = shark_tank_df[shark_columns]

# Standardize the features
scaler = StandardScaler()
scaled_shark_features = scaler.fit_transform(shark_df.fillna(0))

# Apply K-Means Clustering
kmeans_shark = KMeans(n_clusters=5, random_state=42)
kmeans_shark.fit(scaled_shark_features)
shark_tank_df['Shark Cluster'] = kmeans_shark.labels_

# Visualize the clusters using PCA
pca_shark = PCA(n_components=2)
pca_shark_components = pca_shark.fit_transform(scaled_shark_features)

# Plot PCA components with clusters
plt.figure(figsize=(10, 8))
sns.scatterplot(x=pca_shark_components[:, 0], y=pca_shark_components[:, 1], hue=shark_tank_df['Shark Cluster'], palette='Set2', s=100)
plt.title('Shark Clustering (PCA)')
plt.xlabel('PCA Component 1')
plt.ylabel('PCA Component 2')
plt.show()

# Visualize clusters in terms of original features
plt.figure(figsize=(10, 6))
shark_investments = shark_tank_df.melt(id_vars=['Shark Cluster'], value_vars=shark_columns, var_name='Shark', value_name='Investment Amount')
sns.boxplot(x='Shark', y='Investment Amount', hue='Shark Cluster', data=shark_investments, palette='Set2')
plt.title('Shark Investment Amounts by Cluster')
plt.xticks(rotation=90)
plt.show()
```
