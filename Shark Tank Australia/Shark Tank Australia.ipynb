## 3. Data Preprocessing
- **Handling Missing Values**: All missing values in the dataset are dropped to ensure clean data for analysis.
- **Categorical to Numerical Conversion**: Categorical variables such as 'category' and 'investors' are converted to numerical values using `LabelEncoder`.
- **Feature Scaling**: Numerical features like 'amount_requested' and 'amount_offered' are scaled using `StandardScaler` to standardize the data and improve the performance of subsequent analysis techniques.

```python
# List of columns with constant values
constant_columns = [
    'Royalty Deal', 'Deal Has Conditions', 'Steve Present',
    'Janine Present', 'Andrew Present', 'Naomi Present', 'Glen Present', 'Advisory Shares Equity'
]

# Remove constant columns from the DataFrame
shark_tank_df.drop(columns=constant_columns, inplace=True)

# Display the remaining columns in the DataFrame
print("Remaining columns in the DataFrame after removing constant columns:")
print(shark_tank_df.columns)
```

```python
# Display the remaining columns in the DataFrame
print("Remaining columns in the DataFrame:")
print(shark_tank_df.columns)
```

```python
# Checking for missing values in each column
print("Count of missing values in each column before replacement:")
print(shark_tank_df.isnull().sum())
```

```python
# Replace missing values in numerical columns with the mean
numerical_columns = shark_tank_df.select_dtypes(include=[np.number]).columns
for column in numerical_columns:
    shark_tank_df[column].fillna(shark_tank_df[column].mean(), inplace=True)

# Replace missing values in categorical columns with the mode
categorical_columns = shark_tank_df.select_dtypes(include=[object]).columns
for column in categorical_columns:
    shark_tank_df[column].fillna(shark_tank_df[column].mode()[0], inplace=True)

# Checking for missing values in each column after replacement
print("Count of missing values in each column after replacement:")
print(shark_tank_df.isnull().sum())
```

```python
# Filter the DataFrame to include only the numerical columns
shark_tank_numerical_df = shark_tank_df[numerical_columns]
# Calculate the interquartile range (IQR) for each numerical column
Q1 = shark_tank_numerical_df.quantile(0.25)
Q3 = shark_tank_numerical_df.quantile(0.75)
IQR = Q3 - Q1
# Determine the threshold for outliers
outlier_threshold = 1.5
# Identify rows containing outliers in any numerical column
outliers = ((shark_tank_numerical_df < (Q1 - outlier_threshold * IQR)) | (shark_tank_numerical_df > (Q3 + outlier_threshold * IQR)))
# Count the number of outliers in each numerical column
outliers_count=outliers.sum()

print("Outliers count in each numerical columns:")
print(outliers_count)
```

```python
# Remove rows containing outliers
cleaned_shark_tank_df = shark_tank_df[~outliers.any(axis=1)]

# Display the shape of the original and cleaned DataFrames
print("Original DataFrame shape:", shark_tank_df.shape)
print("Cleaned DataFrame shape:", cleaned_shark_tank_df.shape)
```

## 5. Clustering Analysis
- **K-Means Clustering**: K-Means clustering is performed with 3 clusters to group similar funding requests.
- **Visualization**: A scatter plot is created to visualize the clusters based on 'amount_requested' and 'amount_offered', showing how funding requests are grouped.

```python
# Vectorize descriptions using TF-IDF
tfidf_vectorizer = TfidfVectorizer(stop_words='english')
tfidf_matrix = tfidf_vectorizer.fit_transform(shark_tank_df['Business Description'])

# Convert TF-IDF matrix to DataFrame
tfidf_df = pd.DataFrame(tfidf_matrix.toarray(), index=shark_tank_df.index)

# Select numeric columns for clustering
numeric_columns = ['Total Deal Amount', 'Total Deal Equity', 'Received Offer']
numeric_df = shark_tank_df[numeric_columns]

# Combine TF-IDF features with numeric features
combined_df = pd.concat([tfidf_df, numeric_df], axis=1)

# Convert all column names to strings
combined_df.columns = combined_df.columns.astype(str)

# Standardize the combined features
scaler = StandardScaler()
scaled_features = scaler.fit_transform(combined_df)

# Apply K-Means Clustering
kmeans = KMeans(n_clusters=5, random_state=42, n_init=10)
kmeans.fit(scaled_features)
shark_tank_df['Cluster'] = kmeans.labels_

# Map cluster labels to shark names
shark_names = {0: 'Steve', 1: 'Janine', 2: 'Andrew', 3: 'Naomi', 4: 'Glen'}
shark_tank_df['Shark Name'] = shark_tank_df['Cluster'].map(shark_names)

# Visualize the clusters using PCA
pca = PCA(n_components=2)
pca_components = pca.fit_transform(scaled_features)

# Plot PCA components with clusters
plt.figure(figsize=(10, 8))
sns.scatterplot(x=pca_components[:, 0], y=pca_components[:, 1], hue=shark_tank_df['Shark Name'], palette='Set1', s=100)
plt.title('Startup Clustering (PCA)')
plt.xlabel('Business Description Features')
plt.ylabel('Deal and Equity Features')
plt.legend(title='Shark', bbox_to_anchor=(1.05, 1), loc='upper left')
plt.show()

# Visualize clusters in terms of original features
plt.figure(figsize=(10, 6))
sns.countplot(x='Industry', hue='Shark Name', data=shark_tank_df, palette='Set1')
plt.title('Number of Startups per Industry by Shark')
plt.xticks(rotation=90)
plt.legend(title='Shark', bbox_to_anchor=(1.05, 1), loc='upper left')
plt.show()
```

```python
# Select relevant features for shark clustering
shark_columns = [
    'Steve Investment Amount', 'Janine Investment Amount', 'Andrew Investment Amount',
    'Naomi Investment Amount', 'Glen Investment Amount', 'Guest Investment Amount',
    'Steve Investment Equity', 'Janine Investment Equity', 'Andrew Investment Equity',
    'Naomi Investment Equity', 'Glen Investment Equity', 'Guest Investment Equity'
]

shark_df = shark_tank_df[shark_columns]

# Standardize the features
scaler = StandardScaler()
scaled_shark_features = scaler.fit_transform(shark_df.fillna(0))

# Apply K-Means Clustering
kmeans_shark = KMeans(n_clusters=5, random_state=42, n_init=10)
kmeans_shark.fit(scaled_shark_features)
shark_tank_df['Shark Cluster'] = kmeans_shark.labels_

# Map cluster labels to shark names
shark_names = {0: 'Steve', 1: 'Janine', 2: 'Andrew', 3: 'Naomi', 4: 'Glen'}
shark_tank_df['Shark Name'] = shark_tank_df['Shark Cluster'].map(shark_names)

# Visualize the clusters using PCA
pca_shark = PCA(n_components=2)
pca_shark_components = pca_shark.fit_transform(scaled_shark_features)

# Plot PCA components with clusters
plt.figure(figsize=(10, 8))
sns.scatterplot(x=pca_shark_components[:, 0], y=pca_shark_components[:, 1], hue=shark_tank_df['Shark Name'], palette='Set2', s=100)
plt.title('Shark Clustering (PCA)')
plt.xlabel('Total Investment Amount')
plt.ylabel('Total Investment Equity')
plt.legend(title='Shark', bbox_to_anchor=(1.05, 1), loc='upper left')
plt.show()

# Visualize clusters in terms of original features
plt.figure(figsize=(10, 6))
shark_investments = shark_tank_df.melt(id_vars=['Shark Name'], value_vars=shark_columns, var_name='Shark', value_name='Investment Amount')
sns.boxplot(x='Shark', y='Investment Amount', hue='Shark Name', data=shark_investments, palette='Set2')
plt.title('Shark Investment Amounts by Cluster')
plt.xticks(rotation=90)
plt.legend(title='Shark', bbox_to_anchor=(1.05, 1), loc='upper left')
plt.show()
```
